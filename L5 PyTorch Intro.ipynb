```python
import torch
import numpy as np
```

```python
torch.manual_seed(42)
X = 2 * torch.rand(100)
y = 4 + 3 * X + torch.randn(100) * 0.5
```

```python
def compute_loss(X, y, w, b):
  y_pred = X * w + b

  return torch.mean((y_pred - y) ** 2)
```

```python
w = torch.tensor([-20.0], requires_grad=True)
b = torch.tensor([42.0], requires_grad=True)
gamma = 0.1

for epoch in range(300):
  loss = compute_loss(X, y, w, b)
  loss.backward() # PyTorch feature, i don't have to compute derivatives manually

  with torch.no_grad():
    # Use those derivatives here
    w -= gamma * w.grad
    b -= gamma * b.grad

  if epoch % 30 == 0 or epoch == 299:
    print(f"Epoch {epoch}")
    print(f"w = {w.item():.4f}, b = {b.item():.4f}")
    print(f"Loss = {compute_loss(X, y, w, b):.4f}")

  # Reset and start freshly
  w.grad.zero_()
  b.grad.zero_()
```
