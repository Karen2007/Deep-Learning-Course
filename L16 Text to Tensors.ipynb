This lesson is about how words are transformed into numerical data, that the computer can process, what goes in, what vocabularies and embedding tables are, etc.

```python
import torch
from torch import nn
```

```python
text = "The cat didn't sit on the mat!"
```

```python
# A simple tokenizer function, to clean up text before transforming into numbers

def tokenize(text):
  text = text.lower() # Lowercase
  text = text.replace("'", " '").replace("?", " ?").replace("!", " !").replace(".", " .") # Separate punctuation
  text = text.split() # Split into tokens
  return text
```

```python
tokenize(text)
```

```python
# Create an object that maps words to numbers
class Vocabulary():
  def __init__(self):
    # Create a dict that maps tokens to their indices
    self.wordmap = {'<pad>' : 0, # <pad> isn't useful yet
                     '<unk>' : 1}

  # Build vocabulary based on sentences given
  def build(self, sentences):
    for sentence in sentences:
      sentence = tokenize(sentence)
      for word in sentence:
        if word not in self.wordmap.keys():
          self.wordmap[word] = max(self.wordmap.values()) + 1


  # Convert given text to indices of pre-built text
  def to_indices(self, text):
    indices = []

    for word in tokenize(text):
      if word not in self.wordmap.keys():
        indices.append(1)
      else:
        indices.append(self.wordmap[word])

    return indices

```

```python
sentences = [
    "The movie was great!",
    "I hated this terrible film.",
    "What a wonderful experience."
]

vocab = Vocabulary() # Create the new vocab obj
vocab.build(sentences) # Create a mapping of tokens to indices
print(vocab.to_indices("The movie was awful!"))
print(vocab.wordmap)
```

```python
embedding = nn.Embedding(num_embeddings=len(vocab.wordmap), embedding_dim=50) # Create an embedding table for each word in our vocab, each being 50D
indices = torch.tensor(vocab.to_indices("The movie was great!"))
vectors = embedding(indices)
print(vectors.shape)
```
