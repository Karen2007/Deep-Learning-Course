```python
import torch
from torch import nn
```

```python
text = "The cat didn't sit on the mat!"
```

```python
def tokenize(text):
  text = text.lower() # Lowercase
  text = text.replace("'", " '").replace("?", " ?").replace("!", " !").replace(".", " .") # Separate punctuation
  text = text.split() # Split into tokens
  return text
```

```python
tokenize(text)
```

```python
# Create an object that maps words to numbers
class Vocabulary():
  def __init__(self):
    self.word2idx = {'<pad>' : 0,
                     '<unk>' : 1}

  # Create the mapping dict
  def build(self, sentences):
    for sentence in sentences:
      sentence = tokenize(sentence)
      for word in sentence:
        if word not in self.word2idx.keys():
          self.word2idx[word] = max(self.word2idx.values()) + 1
    return None

  # Convert given text to indices of pre-built text
  def to_indices(self, text):
    indices = []

    for word in tokenize(text):
      if word not in self.word2idx.keys():
        indices.append(1)
      else:
        indices.append(self.word2idx[word])

    return indices

```

```python
sentences = [
    "The movie was great!",
    "I hated this terrible film.",
    "What a wonderful experience."
]

vocab = Vocabulary()
vocab.build(sentences)
print(vocab.to_indices("The movie was awful!"))
print(vocab.word2idx)
```

```python
embedding = nn.Embedding(num_embeddings=len(vocab.word2idx), embedding_dim=50)
indices = torch.tensor(vocab.to_indices("The movie was great!"))
vectors = embedding(indices)
print(vectors.shape)
```
