```python
import torch
import matplotlib.pyplot as plt
from torch import nn, optim

torch.manual_seed(42)
X = torch.linspace(-3, 3, 300).reshape(-1, 1)
y = torch.sin(X) + torch.randn(300, 1) * 0.15

# First model with 1 layer of 64 neurons
model1 = nn.Sequential(
    nn.Linear(1, 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)

loss_function = nn.MSELoss()
opt = optim.SGD(model1.parameters(), lr=0.05)

losses = []

for epoch in range(500):
  y_pred = model1(X)
  loss = loss_function(y_pred, y)
  loss.backward()

  opt.step()
  opt.zero_grad()

  if epoch % 50 == 0 or epoch == 499:
    print(f"Epoch {epoch}")
    print(f"Loss = {loss.item():.4f}")

  losses.append(loss.item())

plt.plot(losses)
plt.title('Loss over time of the first NN')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# Second model with 8 hidden layers, 64 neurons each
model2 = nn.Sequential(
    nn.Linear(1, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)

loss_function2 = nn.MSELoss()
opt2 = optim.SGD(model2.parameters(), lr=0.05)

losses2 = []

for epoch in range(500):
  y_pred = model2(X)
  loss2 = loss_function2(y_pred, y)
  loss2.backward()

  opt2.step()
  opt2.zero_grad()

  if epoch % 50 == 0 or epoch == 499:
    print(f"Epoch {epoch}")
    print(f"Loss = {loss2.item():.4f}")

  losses2.append(loss2.item())

plt.plot(losses2)
plt.title('Loss over time of the second NN')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# Same model, different optimizer
model3 = nn.Sequential(
    nn.Linear(1, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)

loss_function3 = nn.MSELoss()
opt3 = optim.Adam(model3.parameters(), lr=0.05)

losses3 = []

for epoch in range(500):
  y_pred = model3(X)
  loss3 = loss_function3(y_pred, y)
  loss3.backward()

  opt3.step()
  opt3.zero_grad()

  if epoch % 50 == 0 or epoch == 499:
    print(f"Epoch {epoch}")
    print(f"Loss = {loss3.item():.4f}")

  losses3.append(loss3.item())

plt.plot(losses3)
plt.title('Loss over time of the second NN using Adam Optimizer')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

As we can see, Adam's optimizer converges quicker than SGD.
