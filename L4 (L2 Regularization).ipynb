```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.randn(200, 3)
y = 5 + 2*X[:, 0] - 3*X[:, 1] + 0.5*X[:, 2] + np.random.randn(200) * 0.5
```

```python
def compute_loss(X, y, w, b, lam):
  y_pred = X @ w + b
  mse = np.mean((y_pred  - y) ** 2)
  penalty = lam * np.sum(w ** 2)
  loss = mse + penalty

  return loss

def compute_gradients(X, y, w, b, lam):
  n = len(X)
  y_pred = X @ w + b
  error = y_pred - y

  dw = (2 / n) * (X.T @ error) + 2 * lam * w
  db = (2 / n) * np.sum(error)

  return dw, db
```

```python
w = np.array([500, 25, -50])
b = 4
gamma = 0.05
lam = 0.1

def perform_descent(X, y, w, b, lam):

  steps_w = [w]
  steps_b = [b]
  losses = [compute_loss(X, y, w, b, lam)]

  print(f"Performing descent with lambda = {lam}\n")
  for epoch in range(1000):

    dw, db = compute_gradients(X, y, w, b, lam)
    w = w - gamma * dw
    b = b - gamma * db

    loss = compute_loss(X, y, w, b, lam)
    steps_w.append(w)
    steps_b.append(b)
    losses.append(loss)

    if epoch % 100 == 0 or epoch == 999:
      print(f"Epoch {epoch}")
      print(f"Current weights: {w}, b = {b:.4f}")
      print(f"Current Loss: {loss:.4f}")

  plt.plot(range(1001), losses, color='green')
  plt.title(f"L2 Gradient Descent with λ={lam}")
  plt.show()

  print("\n\n\n\n")
  return w, b

weights_and_biases = []

for lam in [-0.1, 0, 0.01, 0.1, 10]:
  w, b = perform_descent(X, y, w, b ,lam)
  weights_and_biases.append(f"After descent with lambda = {lam}: w = {w}, b = {b:.4f}, loss = {compute_loss(X, y, w, b, lam):.4f}")
  # Reset the variables
  w = np.array([500, 25, -50])
  b = 4

for i in weights_and_biases:
  print(i)
```
